# Getting Data Ready

```python
# Standard imports
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
heart_disease = pd.read_csv('../../Datas/heart-disease.csv')
heart_disease.head()
```

```python
# Splitting data
X = heart_disease.drop('target', axis = 1)
X.head()
```

```python
y = heart_disease['target']
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2)
```

```python
X_train.shape, X_test.shape
```

```python
y_train.shape, y_test.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier()
clf.fit(X_train, y_train)
clf.score(X_test, y_test)
```

# Converting non-numeric data into numeric data and fitting the model

```python
car_sales = pd.read_csv('../../Datas/car-sales-extended.csv')
car_sales.head()
```

```python
len(car_sales)
```

```python
car_sales['Doors'].value_counts()
```

```python
car_sales.info()
```

```python
# Splitting data into features and labels
from sklearn.model_selection import train_test_split
X = car_sales.drop('Price', axis = 1)
y = car_sales['Price']
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2)

```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor()
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

```python
car_sales.head()
```

```python
# Model could not fit the categorical data
# To solve that converting the categorical data into numerical data
# Method 1
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
categorical_features = ["Make", "Colour", "Doors"]
one_hot = OneHotEncoder()
transformer = ColumnTransformer([('one_hot', one_hot, categorical_features)], remainder = "passthrough")
transformed_X = transformer.fit_transform(X)
```

```python
transformed_X
```

```python
pd.DataFrame(transformed_X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(transformed_X,y, test_size = 0.2)
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

```python
# Another way of converting the categorical values into numeric data
dummies = pd.get_dummies(car_sales, columns = ['Make', 'Colour', 'Doors'])
dummies.head()
```

```python
X, y = dummies.drop('Price', axis = 1), dummies['Price']
X_train, X_test, y_train, y_test = train_test_split(X,y)
model.fit(X_train, y_train)
model.score(X_test, y_test)
```
